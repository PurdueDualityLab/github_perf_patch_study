```python
import pandas as pd
```

```python
df_big = pd.read_csv("/Users/antoniozhong/Documents/dev/purdue/MSR2026/github_perf_patch_study/datasets/ai_perf_prs_with_gpt_analysis.csv")
df_mini = pd.read_csv("/Users/antoniozhong/Documents/dev/purdue/MSR2026/github_perf_patch_study/datasets/ai_perf_prs_with_gpt_analysis_gpt5mini.csv")
```

```python
key = "id"

merged = df_big[[key, "optimization_pattern", "optimization_subpattern"]].merge(
    df_mini[[key, "optimization_pattern", "optimization_subpattern"]],
    on=key,
    suffixes=("_big", "_mini")
)

merged["pattern_match"] = merged["optimization_pattern_big"] == merged["optimization_pattern_mini"]
merged["subpattern_match"] = merged["optimization_subpattern_big"] == merged["optimization_subpattern_mini"]

summary = {
    "total": len(merged),
    "pattern_match_count": int(merged["pattern_match"].sum()),
    "subpattern_match_count": int(merged["subpattern_match"].sum()),
}

summary
```

```python
import pandas as pd
import matplotlib.pyplot as plt

path = "/Users/antoniozhong/Documents/dev/purdue/MSR2026/github_perf_patch_study/datasets/cnn/human_perf_pr_ccn_deltas.csv"
df = pd.read_csv(path)

ccn_df = df[df["metric"] == "AvgCCN"].copy()

plt.figure(figsize=(12, 6))
plt.hist(ccn_df["delta_pct"].dropna(), bins=30)
plt.xlabel("Delta % in AvgCCN (optimized vs original)")
plt.ylabel("Number of PRs")
plt.title("Distribution of AvgCCN Percentage Change for Human Perf PRs")
plt.tight_layout()
plt.show()
```
